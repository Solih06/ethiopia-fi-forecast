This notebook runs the exploratory data analysis pipeline, checks data quality across the unified schema, and visualizes historical access and usage trajectories.

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure local src modules can be resolved cleanly
sys.path.append(os.path.abspath('../src'))
from data_loader import UnifiedDataLoader

loader = UnifiedDataLoader(data_dir='../data/raw')
df = loader.load_raw_data()
obs, envs, links, targets = loader.split_by_record_type(df)

print(f'Observations: {len(obs)}, Events: {len(envs)}, Links: {len(links)}')
```

### Plotting Historical Account Ownership (Access Trajectory)

```python
acc_df = loader.get_indicator_subset(obs, 'ACC_OWNERSHIP')
if not acc_df.empty:
    plt.figure(figsize=(9, 5))
    plt.plot(acc_df['observation_date'], acc_df['value_numeric'], marker='o', color='#1f77b4', linewidth=2)
    plt.title('Ethiopia Account Ownership Growth Engine', fontsize=14)
    plt.ylabel('Percentage (%)')
    plt.ylim(0, 100)
    plt.grid(True, linestyle='--')
    plt.show()
else:
    print('ACC_OWNERSHIP tracking column empty or unmapped.')
```
